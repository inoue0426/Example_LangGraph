```python
df = pd.read_csv('../GraphPINE/data/PC14.All.hgnc.txt.sif.gz', sep='\t', header=None)
df
```

```python
df[df[0].str.contains('TOP2')]
```

```python
graph = NetworkxEntityGraph()
for id, row in df.iterrows():
    graph._graph.add_edge(
        row[0],
        row[2],
        relation=row[1]
    )
```

```python
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(
    model_name="gpt-3.5-turbo-instruct",    
)
chain = GraphQAChain.from_llm(
    llm=llm,
    graph=graph,
    verbose=True
)
```

```python
question = """ TOP1 has the connection with TOP2A ?"""
chain.invoke(question)
```
